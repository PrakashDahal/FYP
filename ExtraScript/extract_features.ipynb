# Extraction of sad features from Audio and storing data

```python
# loading all sad audio and appending features row in csv

fileName1 = 'features'
output = 'sad'

create_file(fileName1)

for i in range(15):
    if i>8:
        y, sr = librosa.load(str("audio_files/DC/sa")+str(i+1)+str(".wav"))
        y1, sr1 = librosa.load(str("audio_files/JE/sa")+str(i+1)+str(".wav"))
        y2, sr2 = librosa.load(str("audio_files/JK/sa")+str(i+1)+str(".wav"))
        y3, sr3 = librosa.load(str("audio_files/KL/sa")+str(i+1)+str(".wav"))
    else:
        y, sr = librosa.load(str("audio_files/DC/sa0")+str(i+1)+str(".wav"))
        y1, sr1 = librosa.load(str("audio_files/JE/sa0")+str(i+1)+str(".wav"))
        y2, sr2 = librosa.load(str("audio_files/JK/sa0")+str(i+1)+str(".wav"))
        y3, sr3 = librosa.load(str("audio_files/KL/sa0")+str(i+1)+str(".wav"))
        
    MFCC0 = list(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13))
    MFCC1 = list(librosa.feature.mfcc(y=y1, sr=sr1, n_mfcc=13))
    MFCC2 = list(librosa.feature.mfcc(y=y2, sr=sr2, n_mfcc=13))
    MFCC3 = list(librosa.feature.mfcc(y=y3, sr=sr3, n_mfcc=13))
    
    mfcc_0 = []
    for i in MFCC0:
        mfcc_0.append(list(i))
        
    mfcc_1 = []
    for i in MFCC1:
        mfcc_1.append(list(i))
        
    mfcc_2 = []
    for i in MFCC2:
        mfcc_2.append(list(i))
        
    mfcc_3 = []
    for i in MFCC3:
        mfcc_3.append(list(i))
   

    
    c = librosa.cqt(y,sr)
    c1 = librosa.cqt(y1,sr1)
    c2 = librosa.cqt(y2,sr2)
    c3 = librosa.cqt(y3,sr3)
    
    chroma0 = list(librosa.feature.chroma_cqt(C=c, sr=sr))  
    chroma1 = list(librosa.feature.chroma_cqt(C=c1, sr=sr1))
    chroma2 = list(librosa.feature.chroma_cqt(C=c2, sr=sr2))
    chroma3 = list(librosa.feature.chroma_cqt(C=c3, sr=sr3))
    
    chroma_0 = []
    leng0 =  len(chroma0)
    for i in range(leng0):
        c0 = []
        for p in chroma0[i]:
            c0.append((p.real, p.imag))
        chroma_0.append(c0)
            
        
    chroma_1 = []
    leng1 =  len(chroma1)
    for i in range(leng1):
        c1 = []
        for p in chroma1[i]:
            c1.append((p.real, p.imag))
        chroma_1.append(c1)
        
    chroma_2 = []
    leng2 =  len(chroma2)
    for i in range(leng2):
        c2 = []
        for p in chroma2[i]:
            c2.append((p.real, p.imag))
        chroma_2.append(c2)
        
    chroma_3 = []
    leng3=  len(chroma3)
    for i in range(leng3):
        c3 = []
        for p in chroma3[i]:
            c3.append((p.real, p.imag))
        chroma_3.append(c3)
        
    append_row(fileName1, mfcc_0, chroma_0, output)
    append_row(fileName1, mfcc_1, chroma_1, output)
    append_row(fileName1, mfcc_2, chroma_2, output)
    append_row(fileName1, mfcc_3, chroma_3, output)
    
    
```

```python
def create_file(fileName):
    with open('extracted_feature/'+'/'+str(fileName)+'.csv', mode='w', newline='') as csv_file:
        fieldnames = ['MFCC_1_m', 'MFCC_2_m', 'MFCC_3_m','MFCC_4_m','MFCC_5_m','MFCC_6_m',
                      'MFCC_7_m','MFCC_8_m','MFCC_9_m','MFCC_10_m','MFCC_11_m','MFCC_12_m','MFCC_13_m',
                      'MFCC_1_v', 'MFCC_2_v', 'MFCC_3_v','MFCC_4_v','MFCC_5_v','MFCC_6_v',
                      'MFCC_7_v','MFCC_8_v','MFCC_9_v','MFCC_10_v','MFCC_11_v','MFCC_12_v','MFCC_13_v','OUTPUT']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        
# appending mfcc row in created csv file
def append_row(fileName, MFCC_M, MFCC_V, output):
    with open('extracted_feature/' + '/'+str(fileName)+'.csv', mode='a', newline='') as csv_file:
        fieldnames = ['MFCC_1_m', 'MFCC_2_m', 'MFCC_3_m','MFCC_4_m','MFCC_5_m','MFCC_6_m',
                      'MFCC_7_m','MFCC_8_m','MFCC_9_m','MFCC_10_m','MFCC_11_m','MFCC_12_m','MFCC_13_m',
                      'MFCC_1_v', 'MFCC_2_v', 'MFCC_3_v','MFCC_4_v','MFCC_5_v','MFCC_6_v',
                      'MFCC_7_v','MFCC_8_v','MFCC_9_v','MFCC_10_v','MFCC_11_v','MFCC_12_v','MFCC_13_v','OUTPUT']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writerow({'MFCC_1_m': MFCC_M[0], 'MFCC_2_m': MFCC_M[1], 'MFCC_3_m': MFCC_M[2],'MFCC_4_m': MFCC_M[3], 'MFCC_5_m': MFCC_M[4],
                         'MFCC_6_m': MFCC_M[5],'MFCC_7_m': MFCC_M[6], 'MFCC_8_m': MFCC_M[7], 'MFCC_9_m': MFCC_M[8],'MFCC_10_m': MFCC_M[9], 
                         'MFCC_11_m': MFCC_M[10], 'MFCC_12_m': MFCC_M[11], 'MFCC_13_m': MFCC_M[12],
                         'MFCC_1_v': MFCC_V[0], 'MFCC_2_v': MFCC_V[1], 'MFCC_3_v': MFCC_V[2],'MFCC_4_v': MFCC_V[3], 'MFCC_5_v': MFCC_V[4],
                         'MFCC_6_v': MFCC_V[5],'MFCC_7_v': MFCC_V[6], 'MFCC_8_v': MFCC_V[7], 'MFCC_9_v': MFCC_V[8],'MFCC_10_v': MFCC_V[9], 
                         'MFCC_11_v': MFCC_V[10], 'MFCC_12_v': MFCC_V[11], 'MFCC_13_v': MFCC_V[12],'OUTPUT':output})
        
fileName3 = 'mean_variance'
    
create_file(fileName3)
```
